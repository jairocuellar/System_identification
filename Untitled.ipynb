```python
import struct

with open('data/s02.dat', 'rb') as f:
    contenido = f.read()

numeros = []
for i in range(0, len(contenido), 4):
    numero = struct.unpack('f', contenido[i:i+4])[0]
    numeros.append(numero)

print(numeros)
```
